## Making Tidy Data by melt, pivot

```python
df2 = pd.DataFrame({'A': ['a','b','c'],
                    'B': [1,3,5],
                    'C': [2,4,6]})
df2
```

```python
pd.melt(df2, id_vars=['A'], value_vars=['B'])
```

```python
pd.melt(df2, id_vars=['A'], value_vars=['B','C'])
```

```python
pd.melt(df2, value_vars=['A','B','C'])
```

```python
pd.melt(df2, value_vars=['A','B','C']).rename(columns={'variable' : 'var',
                                                        'value' : 'val'})
```

```python
df3 = pd.DataFrame({'A': ['a','b','c'],
                    'B': [1,3,5],
                    'C': [2,4,6]})
df3
```

```python
df3 = pd.DataFrame({'bar': ['A', 'B', 'C', 'A', 'B', 'C'],
                    'baz': [1, 2, 3, 4, 5, 6],
                    'foo': ['one', 'one', 'one', 'two', 'two','two']})
df3
```

```python
df3.pivot(index='foo', columns='bar', values='baz')
```

```python
df4 = df3.pivot(index='foo', columns='bar', values='baz').reset_index()
```

```python
df4
```

```python
pd.melt(df4, id_vars=['foo'], value_vars=['A','B','C'])
```

```python
pd.melt(df4, id_vars=['foo'], value_vars=['A','B','C']).sort_values('value')
```

```python
pd.melt(df4, id_vars=['foo'], value_vars=['A','B','C']).sort_values('value').rename(columns = {'value':'baz'})
```
