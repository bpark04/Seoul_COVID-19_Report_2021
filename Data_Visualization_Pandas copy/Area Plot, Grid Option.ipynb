```python
# https://pandas.pydata.org/pandas-docs/stable/user_guide/visualization.html
import pandas as pd
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.DataFrame(np.random.rand(10, 4), columns=["a", "b", "c", "d"])
df.head()
```

```python
df.plot.area()     # default for NaN is 0, make sure to use dropna() or fillna(), it is CUMULATIVE!!
```

```python
df.plot()
```

```python
df.plot.area(stacked=False)
```

```python
df.plot(grid=True)
```
