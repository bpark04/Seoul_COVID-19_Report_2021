```python
import pandas as pd
import numpy as np
%matplotlib inline
s = pd.Series(np.random.randn(1000),
              index=pd.date_range('1/1/2015', periods=1000))
s.plot()
```

```python
s = s.cumsum()
s.plot()
```

```python
r = s.rolling(window=30)
r.mean()
```

```python
# moving average
s.plot(style='k--')
r.mean().plot(style='k--')
```

```python
df = pd.DataFrame(np.random.randn(1000, 4),
                  index=pd.date_range('1/1/2000', periods=1000),
                  columns=['A', 'B', 'C', 'D'])
df = df.cumsum()
df
```

```python
df.plot()
```

```python
df.rolling(window=30).sum().plot(subplots=True)
```

```python
df.rolling(window=len(df), min_periods=1).mean()[:5].plot()
```

```python
df.expanding(min_periods=1).mean()[:5].plot()
```
